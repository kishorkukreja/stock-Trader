# Use nsepy to Get data from web for each day

```python
from nsepy import get_history
```

```python
from datetime import date
```

```python
data = get_history(symbol="RELIANCE", start=date(2010,1,1), end=date(2018,1,8))
```

```python
data.shape
```

```python
data.head()
```

```python
columns = ['Prev Close','Open','High','Close','Low']
```

```python
stock_data=data[columns]
```

```python
stock_data.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```
